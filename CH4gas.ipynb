```python
temperature = pd.read_csv("D:\\DataAnalytics\\Datasets\\Pollution_Temperature\\cleanTemperatureData.csv")
```

```python
temperature = temperature.loc[:,~temperature.columns.str.contains('Unnamed')]
```

```python
temperature.head()
```

```python
temperature.set_index(['Year'],inplace=True)
```

```python
temperature['Mean'] = temperature.mean(axis=1)
```

```python
temperature.head()
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 10)
temperature.plot(kind='line',x=temperature.index,y='Mean')
```

```python
plt.plot(temperature['Mean'][110:-3])
```

```python
plt.plot(temperature['Mean'][110:-3],ch4gas.loc['mean'][2:-2])# negative relation Pollution and Temperature
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure().gca(projection='3d')

x = temperature.index[110:-3]
y = temperature['Mean'][110:-3]
z = ch4gas.loc['mean'][2:-2]

fig.scatter(x, y, z)

fig.set_xlabel('Year Index')
fig.set_ylabel('Temperature in Fahrenhiet')
fig.set_zlabel('CH4 Pollution in Kt')

plt.show()
```
