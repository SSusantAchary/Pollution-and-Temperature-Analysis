```python
temperature.drop('J-D',axis=1,inplace=True)
```

```python
temperature.drop(['D-N','DJF','MAM','JJA','SON','Year.1'],axis=1,inplace=True)
```

```python
temperature.columns
```

```python
temperature.dropna(how='all',axis=0,inplace=True)
```

```python
temperature = temperature[temperature.Year !='Year']
```

```python
temperature.shape
```

```python
temperature.dtypes
```

```python
temperature['Jan'] = temperature['Jan'].astype(float)
```

```python
temperature['Feb'] = temperature['Feb'].astype(float)
```

```python
temperature.replace(to_replace='',value=0,inplace=True)
```

```python
temperature.tail()
```

```python
temperature[['Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']] = temperature[['Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']].astype(float)
```

```python
temperature.plot(kind='scatter',x='Year',y='Jan')#positive correlation of temperature vs year
```
